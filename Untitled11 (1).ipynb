# IMPORT THE REQUIRED LIBRARY

# LOAD THE TRAIN AND TEST DATASET

```python
train = pd.read_csv('/content/C&T train dataset.csv')
test = pd.read_csv('/content/C&T test dataset.csv')
```

```python
def data_pre(data):
  train["employment_st"] = np.where(train["employment_st"].isna(),train["employment_st"].mode(dropna=True),train["employment_st"])
  train["poi"] = np.where(train["poi"].isna(),train["poi"].median(skipna=True),train["poi"])
  train["gurantors"] = np.where(train["gurantors"].isna(),train["gurantors"].mode(dropna=True),train["gurantors"])
  train["age"] = np.where(train["age"].isna(),train["age"].median(skipna=True),train["age"])
  train["housing_type"] = np.where(train["housing_type"].isna(),train["housing_type"].mode(dropna=True),train["housing_type"])
  return data
```

```python
data_pre(train)
```

```python
def data_prep(data):
  test["employment_st"] = np.where(test["employment_st"].isna(),test["employment_st"].mode(dropna=True),test["employment_st"])
  test["poi"] = np.where(test["poi"].isna(),test["poi"].median(skipna=True),test["poi"])
  test["age"] = np.where(test["age"].isna(),test["age"].median(skipna=True),test["age"])
  test["housing_type"] = np.where(test["housing_type"].isna(),test["housing_type"].mode(dropna=True),test["housing_type"])
  return data
```

# SPILTIING THE DATASET

```python
x = train_enc.drop(["sno","Group_no"],axis=1)
y = train_enc["Group_no"]
```

```python
x_train_enc,x_test_enc,y_train_enc,y_test_enc = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train_enc.shape)
print(y_train_enc.shape)
print(x_test_enc.shape)
print(y_test_enc.shape)
```

# MODEL_EVALUTION

# Logistic_Regression

```python
logreg = LogisticRegression()
logreg.fit(x_train_enc,y_train_enc)
```

```python
y_hat_train_logreg = logreg.predict(x_train_enc)
y_hat_test_logreg = logreg.predict(x_test_enc)
```

```python
def model_eval(actual,predicted):
  confu_matrix=confusion_matrix(actual,predicted)
  class_rep=classification_report(actual,predicted)
  acc_score=accuracy_score(actual,predicted)
  print("the accuarcy of the model is :",round (acc_score,2))
  print(confu_matrix)
  print(class_rep)
```

```python
model_eval(y_train_enc,y_hat_train_logreg)
```

```python
model_eval(y_test_enc,y_hat_test_logreg)
```

# Decision_Tree_Classifier

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train_enc,y_train_enc)
```

```python
y_hat_train_dtree=dtree.predict(x_train_enc)
y_hat_test_dtree=dtree.predict(x_test_enc)
```

```python
model_eval(y_train_enc,y_hat_train_dtree)
```

```python
model_eval(y_test_enc,y_hat_test_dtree)
```

```python
model_eval(y_train_enc,y_hat_train_rf)
model_eval(y_test_enc,y_hat_test_rf)
```
